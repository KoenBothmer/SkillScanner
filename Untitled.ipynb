```python
import pandas as pd
pd.options.plotting.backend = "plotly" #interactive plots will be useful in this context
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pickle
import textwrap
```

```python
df = pd.read_excel('reactions_long.xlsx')
```

```python
df.head()
```

```python
df_help = df[df['Question'].str.contains("help")]
df_help=df
```

```python
for question in df_help.Question.unique():
    df_=df_help[df_help['Question']==question]
    fig=go.FigureWidget(data=[go.Histogram(x=df_['Answer'],
                                            nbinsx=5)],
    layout=go.Layout(xaxis={'range': [0.5,5.5]}))
    split_text = textwrap.wrap(question,100)
    fig.layout.title = '<br>'.join(split_text)
    fig.show()
```
