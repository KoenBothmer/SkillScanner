```python
import plotly.graph_objects as go
from plotly.colors import n_colors
def heatmap(sent, model):
    model = SentenceTransformer(model)
    vector = model.encode(sent)
    fig = go.Figure(go.Heatmap(
        z=[vector[:100],vector[:100]],
        colorscale='RdBu',
        showscale=False
        #zmid=0
        ))

    fig.update_layout(height=30, width=3*300, \
                              #font=dict(size=10),\
                              xaxis=dict(showticklabels=False),
                              yaxis=dict(showticklabels=False),
                              margin=go.layout.Margin(l=0, r=0,b=0, t=0 )
                              )



    return fig
```

```python
models = ['average_word_embeddings_glove.6B.300d','all-distilroberta-v1']
for model in models:
    for sent in ["Teamwork",\
                 "Co-operate",\
                "Programming"]:
        print(sent)
        heatmap(sent, model).show()
```

```python
models = ['average_word_embeddings_glove.6B.300d','all-distilroberta-v1']
for model in models:
    print(model)
    for sent in ["You have scripting experience with Python and or R and SQL",\
                 "At least 2 years of relevant experience coding in Python and SQL",\
                "You are skilled in the communication with stakeholders"]:
        print(sent)
        heatmap(sent, model).show()
```
